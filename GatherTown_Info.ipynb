### 출석체크

하루에 한번, 휴게실에 있는 게시판(노션)을 통해 출석체크를 합니다.

출석체크 방법:  
1. 게더타운 로비(또는 각 기수 별 스터디 룸 내) 중앙 응원문구의 날짜는 매일 00시가 되면 변합니다. 해당 날짜와 함께 아바타를 캡쳐합니다. 조가 있는 경우 조원과 함께 찍어서 올리면 됩니다. 꼭 전원이 모인 상태로 캡쳐하지 않아도 괜찮습니다!
